```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("diabetes_prediction_dataset.csv")
df
```

# Remove Age Outliers

```python
df['age'].value_counts()
```

```python
age_80 = df.loc[df["age"] == 80]
age_not_80 = df.loc[df["age"] != 80]
age_80 = age_80.sample(frac=0.3, replace=True, random_state=1)
df = pd.concat([age_not_80,age_80],axis=0)
```

```python
df['age'].value_counts()
```

```python
df
```

# Remove BMI Outliers

```python
Q1 = df['bmi'].quantile(0.25)
Q3 = df['bmi'].quantile(0.75)
IQR = Q3 - Q1

filter = (df['bmi'] >= Q1 - 1.5 * IQR) & (df['bmi'] <= Q3 + 1.5 *IQR)
df = df.loc[filter]
df
```

# Export

```python
df.to_csv("diabetes_preprocessed.csv", index=False)
```
