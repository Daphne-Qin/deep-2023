```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("diabetes_prediction_dataset.csv")
df
```

```python
# get rid of no info smoking rows
df = df[ (df["smoking_history"] != "No Info") & (df["gender"] != "Other") ]

for col in df.columns:
    print(df.groupby(col).size())
```

```python
pd.DataFrame.to_csv(df, "diabetes_clean.csv")
```

```python
df.groupby("smoking_history").size()
```

```python
df.replace("not current", "former", inplace=True)
df.groupby("smoking_history").size()
```

```python
df_ever_current = df.replace("ever", "current")
df_ever_former = df.replace("ever", "former")
df_ever_excluded = df[ df["smoking_history"] != "ever" ]

df_smoking = {"current": df_ever_current, "former": df_ever_former, "excluded": df_ever_excluded}

for key, value in df_smoking.items():
    print(str("ever=") + key, value.groupby("smoking_history").size())
```

```python
# treat df smoking ever excluded as the clean dataframe
df = df_ever_excluded
df.to_csv("diabetes_clean.csv", index=False)
```
