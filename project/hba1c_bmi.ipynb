```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("diabetes_preprocessed.csv")
df
```

```python
df_hba1c_bmi = df[ ["HbA1c_level", "bmi", "diabetes"] ]
df_hba1c_bmi
```

# Scatter Plots

## Negative Plotted First

```python
# plt.figure(figsize=(10,10))
plt.title("Effects of HbA1c on BMI and Diabetes")
plt.xlabel("HbA1c Level")
plt.ylabel("BMI")

plt.scatter("HbA1c_level", "bmi", data=df[ df_hba1c_bmi["diabetes"] == 0 ], color="tab:blue")
plt.scatter("HbA1c_level", "bmi", data=df[ df_hba1c_bmi["diabetes"] == 1 ], color="tab:orange")
```

## Positive Plotted First

```python
# plt.figure(figsize=(10,10))
plt.title("Effects of HbA1c on BMI and Diabetes (Positive Plotted First)")
plt.xlabel("HbA1c Level")
plt.ylabel("BMI")

plt.scatter("HbA1c_level", "bmi", data=df[ df_hba1c_bmi["diabetes"] == 1 ], color="tab:orange")
plt.scatter("HbA1c_level", "bmi", data=df[ df_hba1c_bmi["diabetes"] == 0 ], color="tab:blue")
```

# Observe Data
There seems to be a clear skew towards higher HbA1c levels leading to higher risk of diabetes, so this is just to make sure there's nothing wrong with the data.

## HbA1c

```python
df_hba1c = df_hba1c_bmi.drop(columns="bmi")
df_hba1c
```

```python
sns.displot(df, x="HbA1c_level", hue="diabetes", discrete=True)
plt.title("HbA1c Relations to Diabetes")
plt.xlabel("HbA1c Level")
plt.ylabel("Frequency")
```

```python
df_hba1c.boxplot(column="HbA1c_level", by="diabetes")
plt.title("HbA1c Relations to Diabetes")
plt.xlabel("HbA1c Level")
plt.ylabel("Frequency")
```
